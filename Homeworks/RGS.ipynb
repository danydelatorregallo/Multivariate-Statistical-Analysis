```python
#%%Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import statsmodels.stats.api as sms
from scipy import stats
from statsmodels.compat import lzip
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#%% Cargar base de datos
#Boston House prices
 data_url = "http://lib.stat.cmu.edu/datasets/boston"
 raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
 data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])

# CRIM - per capita crime rate by town
# ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
# INDUS - proportion of non-retail business acres per town.
# CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
# NOX - nitric oxides concentration (parts per 10 million)
# RM - average number of rooms per dwelling
# AGE - proportion of owner-occupied units built prior to 1940
# DIS - weighted distances to five Boston employment centres
# RAD - index of accessibility to radial highways
# TAX - full-value property-tax rate per $10,000
# PTRATIO - pupil-teacher ratio by town
# B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
# LSTAT - % lower status of the population
# MEDV - Median value of owner-occupied homes in $1000's
```

```python
#%% Selección de Variables
#Variable Independiente
X=data[:,np.newaxis,5]
#Variable Dependiente
Y=raw_df.values[1::2, 2]
```

```python
#%% # Dividir los datos en conjuntos de entrenamiento y prueba
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2)
```

```python
#%%Modelo de regresión Lineal
rls=linear_model.LinearRegression()
modelo=rls.fit(X_train,Y_train)
Y_pred=rls.predict(X_test)
```

```python
#%% Datos de la regresión lineal
print("Coeficientes:", rls.coef_)
print("Intercepto:", rls.intercept_)
print("R^2:", rls.score(X_test, Y_test))
```

```python
#%%Parámetros para prueba de hipótesis B1
error=Y_test-Y_pred
ds_error=error.std()
ds_X=X_test.std()
error_st=ds_error/np.sqrt(506)
t1=rls.coef_/(error_st/ds_X)
print(t1)
```

```python
#%% Parámetros para prueba de hipótesis B0
media_X=X_test.mean()
media_XC=pow(media_X,2)
var_X=X_test.var()
to=rls.intercept_/(error_st*np.sqrt(1+(media_XC/var_X)))
print(to)
```

```python
#%% Gráfica del modelo lineal

plt.scatter(X_test[:, 0], Y_test, label='Valores reales')
plt.plot(X_test[:, 0], Y_pred, color='r', label='Predicciones', linewidth=3)
plt.title('Regresión Lineal')
plt.xlabel('Número promedio de habitaciones')
plt.ylabel('Valor Mediano')
plt.legend()
plt.show()

```

```python
#%% Homocedasticidad
error = Y_test - Y_pred
plt.scatter(Y_pred, error, marker='*', alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Homocedasticidad')
plt.show()
```

```python
#%%Modelo de regresión lineal (statsmodel)

# Agregar una constante para el término independiente (B0)
X_train = sm.add_constant(X_train)

# Crear un modelo de regresión lineal con los datos de entrenamiento
modelo = sm.OLS(Y_train, X_train).fit()

# Obtener los resultados del modelo
print(modelo.summary())

# Ahora puedes usar el modelo para hacer predicciones en los datos de prueba
X_test = sm.add_constant(X_test)
Y_pred = modelo.predict(X_test)

# Imprimir las predicciones en los datos de prueba
print("Predicciones en los datos de prueba:")
print(Y_pred)

'''
Interpretación:
    
R-squared es la proporción de la variabilidad total de la variable dependiente que es 
explicada por el modelo de regresión. Se expresa como un valor entre 0 y 1 (o en porcentaje, de 0% a 100%).
Un valor cercano a 1 indica que el modelo explica una gran parte de la variabilidad en los datos.
Un valor cercano a 0 indica que el modelo apenas explica la variabilidad y, por tanto, 
podría no ser un buen predictor de la variable dependiente F globalmente nuestro modelo es significativo

El F-statistic es una prueba global que evalúa si el modelo de regresión en su conjunto es significativo. 
Es decir, compara el modelo completo (con todos los predictores) 
contra un modelo sin predictores (solo el intercepto).
Hipótesis nula del F-test: Todos los coeficientes de regresión (excepto el intercepto) son iguales a cero, lo que implica que el modelo no mejora la predicción de 
Y respecto a la media.
Un valor F alto y su correspondiente p-value bajo (generalmente menor a 0.05) 
indican que al menos uno de los predictores tiene una relación significativa con la variable 
dependiente, y por ende, el modelo es útil.
Un valor F bajo (con p-value alto) sugiere que el modelo no es significativamente mejor 
que un modelo sin predictores, es decir, que las variables explicativas no aportan 
información relevante para explicar Y.

Prob(Omnibus)
Hipótesis nula: Los residuos tienen una distribución normal.
Si el p-value (Prob(Omnibus)) es mayor a 0.05, no se rechaza la normalidad; 
es decir, no hay evidencia suficiente para decir que los residuos se desvían de la normalidad.
Si el p-value es menor a 0.05, se rechaza la hipótesis nula, sugiriendo que los residuos no son normales.

Durbin-Watson
Qué mide:
La estadística Durbin-Watson evalúa la autocorrelación de primer orden
en los residuos de un modelo de regresión.

Interpretación:

Valor teórico: La escala va de 0 a 4.
Valor cercano a 2: Indica ausencia de autocorrelación.
Valores menores a 2: Sugieren autocorrelación positiva 
(los residuos consecutivos tienden a ser similares).
Valores mayores a 2: Sugieren autocorrelación negativa 
(los residuos consecutivos tienden a alternarse).

Jarque-Bera (JB)
Qué mide:
El test Jarque-Bera evalúa la normalidad de los residuos 
basándose en la asimetría y la curtosis.

Interpretación:

Hipótesis nula: Los residuos siguen una distribución normal.
Un JB alto y un p-value bajo (menor a 0.05) indican que los residuos no son normales.
Un JB bajo y un p-value alto (mayor a 0.05) respaldan la normalidad.
'''
```

```python
#%% Visualizar homocedasticidad 

plt.scatter(Y_pred, error, marker='*', alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Homocedasticidad')
plt.show()
```

```python
#%% Forma Estadística de Homocedasticidad
#Breusch-Pagan
#H0: Homocedasticidad (p>0.05)
#H1: No homocedasticidad (p<0.05)
names=['Lagrange multiplier statistic', 'p-value',
        'f-value', 'f p-value']
test = sms.het_breuschpagan(modelo.resid, X_train)
lzip(names, test)

```

```python
#%% Forma gráfica de la  normalidad de los residuos
plt.figure()
plt.hist(modelo.resid)
plt.show()
```

```python
#%% QQ plot

qq_plot = sm.qqplot(modelo.resid, line='45', fit=True)

# Personalizar la apariencia del QQ plot
plt.title("QQ Plot de los Residuos")
plt.xlabel("Cuantiles Teóricos")
plt.ylabel("Cuantiles de los Residuos")

# Mostrar el gráfico
plt.show()
```

```python
#%% Forma estadística de la normalidda (Shapiro-Wilk)
#Ho: Normalidad (p>0.05)
#H1: No normalidad (p<0.05)
names=[' Statistic', 'p-value']
test=stats.shapiro(modelo.resid)
lzip(names,test)

```
