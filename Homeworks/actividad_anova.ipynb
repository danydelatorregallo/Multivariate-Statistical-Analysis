```python
import pandas as pd
import researchpy as rp
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.stats.multicomp as mc
import matplotlib.pyplot as plt
import pingouin as pg
import seaborn as sns
```

```python
data_dict={'Parcela':['P1,P2,P3 y P4', 'P5,P6,P7 y P8', 'P9,P10,P11 y P12'],
           'Ninguno': [480, 465, 470],
    'Fert_A': [500, 525, 490],
    'Fert_B': [600, 620, 590],
    'Fert_C': [550, 575, 560]
}

df_parcelas = pd.DataFrame(data_dict)
print(df_parcelas)
df_long = pd.melt(df_parcelas, id_vars='Parcela', 
                  value_vars=['Ninguno', 'Fert_A', 'Fert_B', 'Fert_C'],
                  var_name='Tratamiento', value_name='Rendimiento')
print(df_long)
```

```python
# Realizar ANOVA
modelo = ols('Rendimiento ~ C(Tratamiento)', data=df_long).fit()
anova_table = sm.stats.anova_lm(modelo, typ=2)

# Mostrar resultados de ANOVA
print("\nTabla ANOVA:")
print(anova_table)
```

Se rechaza la hipotesis nula 

```python
# Prueba de comparaciones múltiples de Tukey
tukey = mc.MultiComparison(df_long['Rendimiento'], df_long['Tratamiento'])
tukey_resultados = tukey.tukeyhsd()

print("\nPrueba de Tukey (Comparaciones múltiples):")
print(tukey_resultados)
```
