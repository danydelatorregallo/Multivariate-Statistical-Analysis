## _Regresión lineal_

El análisis de regresión tiene como objetico explicar la relación que existe entre una variable dependiente con una o más variables independientes.

**La regresión lineal es útil para predecir una respuesta cuantitativa.**

**La regresión lineal simple ayuda a predecir una respuesta (variable) cuantitativa Y en funcion de una sola variable predictora X. Tambien asume que hay una relación lineal entre X e Y**

### Hipótesis del modelo:
Linealidad: La relacion que tiene X y Y es lineal.

Normalidad: Los errores y las variables siguen una **distribución normal**.

Homocedasticidad: La varianza de los errores es constante, la dispersión de los datos es constante.

Independencia: El error es independiente de X.


![Imagen de WhatsApp 2025-04-01 a las 11.20.02_16ce7374.jpg](<attachment:Imagen de WhatsApp 2025-04-01 a las 11.20.02_16ce7374.jpg>)

### Error estándar:

El error estándar de beta 1 refleja cuan sensible es la estimacion del coeficiente de pendiente a las variaciones aleatorias de la muestra de datos.

Cuanto mas grande sea el error estándar, mayor será la incertidumbre sobre el valor real de Beta1.

Por el contrario, un error estándar pequeño siguere que el estimador es más preciso.

### Interpretación de parámetros estadísticos:
**Criterio Akaike AIC:**
De acuerdo al AIC el modelo que mejor se ajusta a los datos es el que explica la mayor variación usando menos vriables independientes.

**_AIC=2k-2*ln(L)_**

Un AIC más bajo implica un mejor modelo.

Prob(Omnibus): Se utiliza para verificar que los errores se distribuyen normalmente.

H0:Los errores se distribuyen normalmente.

H1:Los errores no se distribuyen normalmente.

**Se espera que Prob(Omnibus) sea cercano a 1.**

Durbin-Watson: Verifica que la varianza de los errores es constante. Se prefiere un valor cercano a 2.

## _Regresión Logística_

La clasificación es el proceso de asignar una o más etiquetas (clases) a cada observación de un conjunto de datos.

En un caso binario solo se asigna como "0" o "1", puede ser "falso" o "positivo", "ocurre" o "no ocurre".

**Por qué centrarse en la Regresión Logística:**

Es un punto de partida natural para problemas de clasificación binaria

**Conexión con la clasificación binaria:**
Define una decisión de clasificación usando un umbral, generalmente 0.5:

Si o(z) >+ 0.5, se clasifica como 1.

Si o(z)<- 0.5, se clasifica como 0.

**ODDS RATIO (Razón de Probabilidades)**

Si odds = 1, la probbilidad de que ocurra el evento es igual a la probabilidad de que no ocurra (p=0.5)

Si odds >+ 1, el evento es más probable (p>+0.5).

Si odds <- 1, el evento es menos probable (p<-0.5)

### **Evaluación del modelo**
**_Accuracy_:**

Mide el porcentaje de predicciones correctas sobre el total de predicciones:

Bueno cuando las clases están balanceadas.

No es confiable si las clases están desequilibradas (ej., si el 95% de los datos son de una sola clase, un modelo que predice siempre esa clase tendrá 95% de accuracy pero no será útil).


**_Precision_:**

Mide cuántas de las predicciones positivas fueron realmente correctas

Útil cuando FP (falsos positivos) son costosos.

**_Recall_:**

Mide cuántas de las instancias positivas reales fueron detectadas correctamente

Útil cuando los FN (falsos negativos) son costosos.

**_F1-Score_:**

Es el balance entre Precision y Recall, usando su media armónica

Útil cuando hay desequilibrio en las clases.

Si F1-Score es alto, significa que el modelo maneja bien tanto FP como FN.

**Curva de Roc**

Traza la tasa de Verdaderos Positivos vs. la Tasa de Falsos Positivos

## **_ANOVA_**
ANOVA= Análisis de la VARIANZA

¿QUÉ ES EL ANOVA?

El análisis de la varianza es una técnica estadística que se utiliza para evaluar si existen diferencias significativas en las medias de tres o más grupos o tratamientos en un conjunto de datos.

**EL ANOVA BUSCA DETERMINAR SI AL MENOS UNO DE LOS GRUPOS ES ESTADÍSTICAMENTE DIFERENTE DE LOS DEMÁS**

**TIPOS DE ANOVA**

ANOVA DE UN FACTOR:

Se compara un solo factor categorico (variable independiente) con tres o más niveles o grupos.

ANOVA DE DOS FACTORES:

Se comparan dos factores categoricos al mismo tiempo. Permite analizar los efectos individuales de cada factor y su interacción en la variable dependiente.

ANOVA DE MEDIAS REPETIDAS:

Combina la idea de un ANOVA de un factor con el enfoque de medidas repetidas.

**SUPUESTOS DEL ANOVA:**
Los supuestos del ANOVA son similares a los de la regresión lineal y están en función a la distribución de los errores y la igualdad de las varianzas.

La variable dependiente debe de medirse al menos a nivel de INTERVALO

Normalidad: Los valores de la muestra provienen de una distribución normal, al igual que sus residuos deben de seguir una distribución normal.

Independencia de las observaciones: Los valores en un grupo o tratamiento no deben de estar relacionados ni influenciados por los valores en otro grupo.

Homocedasticidad: La varianza de los residuos debe ser constante en todos los niveles o tratamientos del factor categórico.

Linealidad: La relacion entre la variable independiente (factor) y la variable dependiente es lineal.

## **Modelo ANOVA**

El ANOVA permite comparar las medias de varias poblaciones a partir del estudio de sus varianzas.

El concepto fundamental del ANOVA es que todo valor observado puede expresarse.

**Comparaciones múltiples (Prueba Post-Hoc):**

La prueba de Anova proporciona solo evidencia de que si existe una diferencia estadisticamente significativa, pero no proporciona evidencia estadistic  sobre que media o medias son diferentes.

Se tiene que usar una prueba **Tukey**

## **_ANCOVA_**

Análisis de la COVARIANZA (ANCOVA)

¿QUÉ ES EL ANCOVA?

Es una técnica estadistica que combina elementos de ANOVA y regresión lineal.

Su objetivo principal es comparar las medias de diferentes grupos ajustando por una o más variables continuas llamadas covariables.

ANCOVA introduce covariables, es decir, factores adicionales que pueden influir que pueden influir en la variable dependiente. 

**DEFINICIÓN**
ANCOVA ajusta los valores medios de los grupos en función de las diferencias en las covariables , permitiendo una comparación más precisa.

**OBJETIVOS**
Eliminar cualquier error que pueda sesgar los resultados.



**SUPUESTOS DEL ANCOVA**

Nivel de medición de la variable dependiente: La  variable dependiente debe ser de al menos nivel de intervalo (como en ANOVA).

NORMALIDAD: 

Los valores de la muestra deben provenir de una distribución normal.

Los residuos (errores del modelo) deben seguir una distribución normal.

INDEPENDENCIA DE LAS OBSERVACIONES: 

Cada observacion debe ser seleccionada de manera aleatoria e independiente de las demás.

HOMOCEDASTICIDAD:

La varianza de los residuos debe ser constante en todos los niveles del factor categórico.

LINEALIDAD:

La relacion entre la covariable y la variable dependiente debe ser lineal.


INDEPENDENCIA ENTRE LA COVARIABLE Y EL FACTOR CATEGÓRICO:

La covariable no debe estar correlacionada con la variable independiente (factor), ya que ANCOVA asume que las diferencias entre grupos no son influenciadas por la covariable.

EFECTO ADITIVO DE LA COVARIABLE:

Se asume que la covariable afecta a la variable dependiente de la misma manera en todos los grupos, es decir, no hay interacción entre la covariable y el factor.

La pendiente de la regresión de la covariable es la misma para todos los niveles del factor

## **_Selección de variables_**



Es un proceso crítico en los campos de la estadítica, análisis de los datos, y ciencia de datos. 

_LA IMPORTANCIA DE LA SELECCIÓN DE VARIABLES_:

REDUCCIÓN DE DIMENSIONALIDAD: Simplifica el modelo al eliminar variables irrelevantes.

MEJORA EL RENDIMIENTO: Menos multicolinealidad conduce a estimaciones mas robustas y precisas.

INTERPRETABILIDAD: Facilita la comprensión de la influencia de cada predictor.

EFICIENCIA COMPUTACIONAL: Disminuye el costo en tiempo y recursos al trabajar con conjuntos de datos grandes.


**Tipos de metodos de seleccion de variables**

Método de filtrado

Métodos de envoltura

Métodos integrados

**MÉTODOS DE FILTRADO**

Pruebas de correlación (coeficiente de Pearson)

Test Chi-cuadrado para variables categoricas

VIF (Factor de Inflación de la Varianza)

**MÉTODOS DE ENVOLTURA**

FORWARD: Comienza sin predictores y va agregando aquellos que mejoran significativamente el modelo.

BACKWARD: Inicia con el modelo completo y elimina variables poco significativas.

STEPWISE: Combina ambos enfoques, permitiendo agregar o eliminar en cada paso.

Ejemplo de criterio- AIC:

**AIC=2Kln(L)**

donde k es el numero de paraetros y L es la funcion de verosimilitud.

Ventajas: Evalúan directamente el impacto de cada variable en el desempeño del modelo.

**_FORWARD_**

En cada etapa se introduce la variable más significativa hasta una cierta regla de parada.

**_BACKWARD_**

En cada etapa se elimina la variable menos influyente, hasta que no procede suprimir ningún término más.

**_STEPWISE_**

Comienza como el de introducción progresiva, pero en cada etapa se plantea si todas las variables introducidas deben de permanecer en el modelo.

**MÉTODOS INTEGRADOS**
Realizan la selección de variables durante el proceso de entrenamiento del modelo.
