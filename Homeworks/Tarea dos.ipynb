```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Cargar los datos
data=pd.read_csv(r"C:\Users\Daniela de la Torre\OneDrive\Escritorio\4to_Semestre\Analisis Estadistico Multivariado\archive (8)\apple_sales_2024.csv")
data
```

Seleccion de variables:

Region: The geographic area where sales were recorded, including North America, Europe, Greater China, Rest of Asia, and Rest of the World.

iPhone Sales: Unit sales of iPhones, measured in millions.

iPad Sales: Unit sales of iPads, measured in millions.

Mac Sales: Unit sales of Mac computers, measured in millions.

Wearables: Unit sales of Apple’s wearable products, such as Apple Watch and AirPods, measured in millions.

Services Revenue: Revenue generated from Apple’s services, such as iCloud, App Store, and Apple Music, measured in billions of dollars.

```python
select_columns= ["State",	"Region",	"iPhone Sales (in million units)",	"iPad Sales (in million units)",	"Mac Sales (in million units)",	"Wearables (in million units)",	"Services Revenue (in billion $)"]
data=data[select_columns]
```

```python
# Resumen estadístico del dataset
print(data.describe())

# Análisis unidimensional: Trip Duration
plt.figure(figsize=(8,5))
sns.histplot(data["iPhone Sales (in million units)"], bins=50, kde=True)
plt.title("Apple iPhone Sales")
plt.xlabel("iPhone Sales")
plt.ylabel("Frecuencia")
plt.show()

# Calcular medidas de tendencia central y dispersión
mean_duration = data["iPhone Sales (in million units)"].mean()
median_duration = data["iPhone Sales (in million units)"].median()
std_duration = data["iPhone Sales (in million units)"].std()

print(f"Media: {mean_duration:.2f} segundos")
print(f"Mediana: {median_duration:.2f} segundos")
print(f"Desviación estándar: {std_duration:.2f} segundos")

# Análisis bidimensional: Correlación entre trip_duration y user_type
plt.figure(figsize=(8,5))
sns.boxplot(x="iPhone Sales (in million units)", y="Services Revenue (in billion $)", data=data)
plt.title("Venta de iPhone respecto a sus servicios")
plt.xlabel("iPhone Sales")
plt.ylabel("Services Revenue")
plt.yscale("log")
plt.show()

# Calcular correlación entre variables numéricas
# Seleccionar solo columnas numéricas
numeric_data = data.select_dtypes(include=[float, int])

# Calcular la matriz de correlación
correlation_matrix = numeric_data.corr()
print("Matriz de correlación:")
print(correlation_matrix)

# Mapa de calor de correlación
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```
